```python
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label = "sin")
plt.plot(x, y2,  linestyle = "--", label = "cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```

```python
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum( x*w ) + b
    if tmp > 0:
        return 1
    elif tmp <= 0:
        return 0
```

```python
print(AND(0, 0))
print(AND(0, 1))
print(AND(1, 0))
print(AND(1, 1))
```

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = np.sum( x*w ) + b
    if tmp > 0:
        return 1
    elif tmp <= 0:
        return 0
```

```python
print(NAND(0, 0))
print(NAND(0, 1))
print(NAND(1, 0))
print(NAND(1, 1))
```

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum( x*w ) + b
    if tmp > 0:
        return 1
    elif tmp <= 0:
        return 0
```

```python
print(OR(0, 0))
print(OR(0, 1))
print(OR(1, 0))
print(OR(1, 1))
```

```python
def XOR(x1, x2):    
    s1 = NAND(x1, x2)    
    s2 = OR(x1, x2)    
    y = AND(s1, s2)    
    return y
```

```python
print(XOR(0, 0))
print(XOR(0, 1))
print(XOR(1, 0))
print(XOR(1, 1))
```
